```python
import matplotlib.pyplot as plt
import json
```

```python
filenames = ['LSTM', 'Random', 'Uniform', 'RL', 'Loser', 'UCRP', 'Winner']
```

```python
for filename in filenames:
    path = filename + ".txt"
    f = open(path, 'r')
    data = json.load(f)
    print(len(data))
    plt.plot(range(len(data)), data, label=filename)

plt.xlabel('Days')
plt.ylabel('Value in US Dollars')
plt.title('Portfolio Performances')
plt.legend()
plt.show()
```
