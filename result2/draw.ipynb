```python
import matplotlib.pyplot as plt
import json
```

```python
filenames = ['LSTM', 'Random', 'Uniform', 'PG', 'Loser', 'UCRP', 'Winner']
# filenames = ['PG']
```

```python
length = 192
times = range(length)

for filename in filenames:
    path = filename + ".txt"
    f = open(path, 'r')
    data = json.load(f)
    print(len(data))
    plt.plot(times, data, label=filename)

plt.xlabel('Days')
plt.ylabel('Value in US Dollars')
plt.title('Portfolio Performances')
plt.legend()
plt.show()
```
